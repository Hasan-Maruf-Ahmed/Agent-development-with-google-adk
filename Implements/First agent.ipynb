```python
from dotenv import load_dotenv
import os

load_dotenv()

try:
    GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
    print("✅ Gemini API key loaded successfully.")
except Exception as e:
    print(f"❌ Authentication Error. Details: {e}")

```

```python
from google.adk.agents import Agent
from google.adk.models.google_llm import Gemini
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print("✅ All imports successful...")
```

```python
retry_config = types.HttpRetryOptions(
    attempts=5,
    exp_base=7,
    initial_delay=1,
    http_status_codes=[429, 500, 503, 504],
)
```

```python
root_agent = Agent(
    name="helpful_assistant",
    model=Gemini(
        model="gemini-2.5-flash-lite",
        retry_options=retry_config,
    ),
    description="A helpful assistant that can answer questions about anything.",
    instruction="Use the tools available to you for the current info or if unsure to answer the user's questions as best as you can.",
    tools=[google_search],
)

print("✅ Agent created successfully...")
```

```python
runner = InMemoryRunner(agent=root_agent, app_name="agents")

print("✅ Runner created.")
```

```python
response = await runner.run_debug(
    "Is there going to be a major earthquake in Bangladesh?",
)
```
